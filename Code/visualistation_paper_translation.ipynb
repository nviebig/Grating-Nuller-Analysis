```python
import numpy as np
import matplotlib.pyplot as plt

# Define the parameters
I_a = 1.0  # Average intensity
I_b = 1  # Modulation amplitude
delta = 1  # Phase offset

# Create a grid of (x, y) points
x = np.linspace(-10, 10, 500)
y = np.linspace(-10, 10, 500)
X, Y = np.meshgrid(x, y)

# Define the phase function (here we use a simple radial pattern for demonstration)
phi = np.sqrt(X**2 + Y**2)

# Calculate the intensity distribution
I = I_a + I_b * np.cos(phi + delta)

# Plot the intensity distribution
plt.figure(figsize=(8, 6))
plt.pcolormesh(X, Y, I, shading='auto', cmap='viridis')
plt.colorbar(label='Intensity')
plt.title('Interference Pattern Intensity Distribution')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
